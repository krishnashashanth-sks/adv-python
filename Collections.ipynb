### Decorators

```python

def repeat(n):
  def myDecorator(func):
    def wrapper():
      print("start")
      for _ in range(n):
        func()
      print("end")
    return wrapper
  return myDecorator
@repeat(n=5)
def dosomething():
  print("SKS")
dosomething()
```

```python
class CountCalls:
  def __init__(self,func,*args,**kwargs):
    self.func=func
    self.num_calls=0
  def __call__(self,*args,**kwargs):
    self.num_calls+=1
    print(f"This is executed {self.num_calls}times")
    return self.func(*args,**kwargs)
@CountCalls
def say_hello():
  print("Hello")
say_hello()
say_hello()
```

### Generators

```python
def generator():
  yield 1
  yield 2
  yield 3
for i in generator():
  print(i)
sorted(generator())
```

```python
### Fibonacci
def Fibonacci(limit):
  a,b=0,1
  while(a<limit):
    yield a
    a,b=b,a+b
for i in Fibonacci(10):
  print(i)
```
